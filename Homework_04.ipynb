## Задача 1

## Задача 3

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
df.describe()
```

```python
df['waterfront'].replace({0: 'no', 1: 'yes'}, inplace=True)
```

```python
fig = plt.subplots(figsize=(16, 6))
ax_31 = sns.boxplot(x=df['price'], y=df["waterfront"], whis=1.5)
ax_31.set_title('Зависимость цены от вида на набережную')
ax_31.set_xlabel('Цена')
ax_31.set_ylabel('Наличие вида на набережную');

tick = mtick.StrMethodFormatter('${x:,.0f}')
ax_31.xaxis.set_major_formatter(tick)
plt.xticks(rotation=30)

plt.show()
```

**Вывод:**  
Представленный график показывает, что все дома, которые имеют вид на море дороже тех, у которых его нет. Также можно отметить, что разброс цен у домов с видом на море намного больше.

```python
data_year = df.groupby('yr_built')['price'].agg(mean='mean').reset_index().sort_values(by='yr_built')
data_year.head()
```

```python
fig, ax_year = plt.subplots(figsize=(10, 4))
ax_year.plot(data_year['yr_built'], data_year['mean'], color='green')
ax_year.set_xlabel('Год постройки')
ax_year.set_ylabel('Средняя цена')
ax_year.set_title('Зависимость средней цены от года постройки')

fmt = '${x: ,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax_year.yaxis.set_major_formatter(tick)

plt.show()
```

**Вывод:**  
На данном графике показано, что, в среднем, цена домов, построенных в 1930-годы и в 2000-е выше. Также дороже стоят дома, построенные в начале прошлого века и есть пик падение - дома, построенные в 1940-е.

```python
fig, ax_sqft = plt.subplots(figsize=(8, 6))
ax_sqft.scatter(df['price'], df['sqft_living'], s=20, alpha=0.5, color='purple', edgecolor='black')
ax_sqft.set_xlabel('Цена')
ax_sqft.set_ylabel('Жилая площадь')
ax_sqft.set_title('Зависимость цен от жилой площади')

fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax_sqft.xaxis.set_major_formatter(tick)
plt.xticks(rotation=30)

plt.show()
```

**Вывод:**  
Данный график демонстрирует, что зависимость между ценой и жилой площадью прямая, при этом наибольшее количество предложений сконцентрированы в районе до 4000 кв.м. и ценой до 2 млн.долл.США.

```python
data_bedr = df.groupby('bedrooms')['price'].agg(mean='mean').reset_index().sort_values(by='bedrooms')
```

```python
fig = plt.subplots(figsize=(8, 6))
ax_bedr = sns.barplot(x=data_bedr['bedrooms'], y=data_bedr['mean'])
ax_bedr.set_title('Зависимость средней цены от количества спален')
ax_bedr.set_xlabel('Количество спален')
ax_bedr.set_ylabel('Средняя цена');

tick = mtick.StrMethodFormatter('${x:,.0f}')
ax_bedr.yaxis.set_major_formatter(tick)

plt.show()
```

**Вывод:**  
На данной гистограмме представлена зависимость средней цены от количества спален в доме. При этом, в среднем, цена растет с ростом числа комнат в доме, но начиная с 9 комнат, стоимость начинает падать.

```python
fig, ax_floors = plt.subplots(figsize=(10, 6))
ax_floors.scatter(df['price'], df['floors'], s=20, alpha=0.5, color='green')
ax_floors.set_xlabel('Цена')
ax_floors.set_ylabel('Количество этажей')
ax_floors.set_title('Зависимость цен от этажности дома')

fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax_floors.xaxis.set_major_formatter(tick)
plt.xticks(rotation=30)

plt.show()
```

**Вывод:**  
Данный график показывает, что самый высокий разброс цен у домов с одним и двумя этажами, при этом, большая часть предложений у домов с любой этажностью до 1 млн. долл. США

Несмотря на то, что порядок цен на недвижимость крайне широкий, абсолютное большинство домов стоят в пределах 500-800 млн.долл.США. При этом, большая часть дорогостоящих домов приходится на дома с видом на набережную, которых представлено крайне мало (менее 1%).

Дома в регионе разнообразные: множество предложений с разными площадями, этажностью, с разным количеством комнат. Тем не менее, можно сделать выводы, что чаще всего на рынке встречаются дома:
